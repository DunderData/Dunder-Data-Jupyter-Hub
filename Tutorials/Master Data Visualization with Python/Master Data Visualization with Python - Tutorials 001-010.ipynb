## Master Data Visualization with Python - Tutorial -1 Explicit with pandas `plot` method


Pandas plot method defaults to a line plot. This isn't obvious. Be explicit and set kind to 'line' to be clear.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('../../mdap.mplstyle')
stocks = pd.read_csv('../data/stocks/stocks10.csv', parse_dates=['date'])
stocks.head(3)
```

## Master Data Visualization with Python - Tutorial -1 Explicit with pandas `plot` method DataFrame Plotting - Explicit and Implicit

* Explicit - provide `x` and `y`
* Implicit - each **column** is plotted independently as the y-values using the **index** as the x-values

```python
stocks.head(3)
```

```python
# explicitly set x and y
stocks.plot(x='date', y=['MSFT', 'FB'], kind='line');
```

```python
# implicit - use index for x values. Each column plotted as its own line
stocks.set_index('date').plot(kind='line');
```
